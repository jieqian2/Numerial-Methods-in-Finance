```python
import pandas as pd
import numpy as np

df_mesh1 = pd.read_csv("mesh1.csv")
df_mesh2 = pd.read_csv("mesh2.csv")
df_mesh3 = pd.read_csv("mesh3.csv")
```

```python
mesh1 = df_mesh1.to_numpy()
mesh2 = df_mesh2.to_numpy()
mesh3 = df_mesh3.to_numpy()
```

```python
len(mesh1[:,:-2][0])
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-2, 3, 0.05)
T = np.arange(0, 1, 0.01)
X, T = np.meshgrid(X, T)
ax.plot_surface(X, T, mesh1[:,:-2], rstride=1, cstride=1, cmap=cm.viridis)
ax.view_init(elev=45,azim=120)
ax.set_xlabel("X")
ax.set_ylabel("T")
ax.set_zlabel("U")
plt.show()
```
